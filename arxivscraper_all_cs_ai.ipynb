```python
import arxivscraper
import pandas as pd
import numpy as np
import time
```

```python
start_dates = pd.date_range('2006-12-31','2017-12-31',freq='Y').strftime('%Y-%m-%d')
end_dates = pd.date_range('2007-12-31','2018-12-31',freq='Y').strftime('%Y-%m-%d')

dates = list(zip(start_dates,end_dates))

for date in dates:
    scraper = arxivscraper.Scraper(category='cs',
                               date_from=date[0],
                               date_until=date[1],
                               filters={'categories':['cs.ai']})
    output = scraper.scrape()
```

```python
dfs = []

for date in dates:
    df = pd.read_pickle('cs_'+date[0]+'_'+date[1])
    dfs.append(df)
    
cs_ai = pd.concat(dfs)

cs_ai.head()
```

```python
for df in dfs:
    print(len(df))
```

```python
len(cs_ai)
```

```python
cs_ai.drop_duplicates('id',inplace=True)

len(cs_ai)
```

```python
cs_ai['year-month'] = pd.to_datetime(cs_ai['created'],format='%Y-%m-%d').dt.strftime('%Y-%m')

cs_ai.head()
```

```python
cs_ai.reset_index(drop=True,inplace=True)
```
